```python
import json

# Convert the string to a Python list
# data_list = json.loads(data_str)
```

```python
def convert(obj):
    L = []
    for i in json.loads(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert2(obj):
    L = []
    counter = 0
    for i in json.loads(obj):
        if counter!=3:
            L.append(i['character'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies['cast']
```

```python
movies['cast'][0]
```

```python
movies['crew']
```

```python
def director(obj):
    L = []
    for i in json.loads(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies['crew']
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'][0]
```

```python
movies['cast']
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
new_df = movies[['id' , 'title' , 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][3]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
import nltk
nltk.download('punkt')
from nltk.stem import PorterStemmer
 
ps = PorterStemmer()
```

```python
def stemm(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stemm)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
vectors = cv.fit_transform(new_df['tags'])
vectors = vectors.toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python

len(cv.get_feature_names_out())
```

```python
for i in cv.get_feature_names_out():
    print(i)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    movie_similarity_list = similarity[movie_index]
    movie_list = sorted(list(enumerate(movie_similarity_list)),reverse=True , key = lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
